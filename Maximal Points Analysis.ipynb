Right to left Variant

```python
comprison_list_rl = []
def maximal_points_rl(data):
    global comprison_list_rl
    sorted_points = sorted(data, key=lambda point: point[0],reverse=True)
    n = len(sorted_points)
    result = [sorted_points[0]]
    i = 1
    # Initialize the comparison counter
    comparison_count = 0
    while  i < n :
        comparison_count += 1
        x, y = sorted_points[i]
        if y > result[-1][1]:
            result.append((x, y))
        i+=1
    comprison_list_rl.append(comparison_count)
    # print("The Number of comparison : ", comparison_count)
    return result
# maximal_points_rl(data)
```

Left to Right Variant

```python
comprison_list_lr = []
def maximal_points_lr(data):
    global comprison_list_lr
    sorted_points = sorted(data, key=lambda point: point[0])
    n = len(sorted_points)
    result = [sorted_points[0]]
    i = 1
    # Initialize the comparison counter
    comparison_count = 0
    while i < n :
        comparison_count += 1
        x, y = sorted_points[i]
        if y > result[-1][1]:
            result.pop()
            j = 0
            while j < len(result):
                comparison_count += 1
                if y > result[-1][1]:
                    result.pop()
                else:
                    break
            result.append((x, y))
        else:
            result.append((x, y))
        i += 1
    comprison_list_lr.append(comparison_count)
    # print("The Number of comparison : ", comparison_count)
    return result
    
# print(maximal_points_lr(data))
```

```python
def generate_random_points(num_points):
    min_coord = 0
    max_coord=num_points
    points = []
    for _ in range(num_points):
        x = random.randint(min_coord, max_coord)
        y = random.randint(min_coord, max_coord)
        points.append((x, y))
    maximal_points_lr(points)
    maximal_points_rl(points)
    return 
```

```python
num = []
for i in range(20,2000,10):
    num.append(i)
    generate_random_points(i)
```

```python
# Plot the data using Matplotlib
plt.scatter(num, comprison_list_lr, label='Left to Right Variant')
plt.scatter(num, comprison_list_rl, label='Right to Left Variant')


plt.xlabel('Number of Input')
plt.ylabel('Number of Comparison')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
